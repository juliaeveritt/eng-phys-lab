```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (4,3)  # good default size for column-width figure
```

```python
filename = 'data/feb16/05501.csv'

df_temp = pd.read_csv(filename, header=None)
df_temp.columns = ['Header', 'Headerdata', 'n1', 't', 'V', 'n2']
df_temp = df_temp.drop(columns=['n1', 'n2'])
```

```python
plt.plot(df_temp['t'], df_temp['V'])
plt.xlabel("Time [s]")
plt.ylabel("Voltage [V]")
```
