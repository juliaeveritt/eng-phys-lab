```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.constants as const
```

```python
# Colour palette
c = ["#006ba6", "#0496ff", "#82a9a1", "#ffbc42", "#ec674e", "#d81159", "#8f2d56"]
```

```python
# Plot settings
plt.rcParams['figure.dpi'] = 300
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (4,3)  # good default size for column-width figure
```

```python
def get_energy(bins):
    return 0.1263 * bins - 0.6
```

## Background readings
* Coarse gain: x10
* Fine gain: just under 12.5
* Integrate: 200 ns
* Diff: 500 ns

```python
bin_arr = []
energy_arr = []
```

```python
filepath = 'data/background_air.csv'
duration = 650  # s

df = pd.read_csv(filepath, header=7)
df.columns = ['ch', 'n']

bg_ch = df['ch']  # bin
bg_n = df['n'] / duration  # counts/s

plt.plot(get_energy(bg_ch), bg_n, 'k')
plt.xlabel('Energy [keV]')
plt.ylabel('Counts/s')
plt.yscale('log')
plt.title('Background')
```

## Spectrum 1: short sampling time
Looking for Pb peak around 351.9 keV (bin 3000) or 77 keV (bin 1000)

```python
filepath = 'data/air_short.csv'
duration = 1238  # s

df = pd.read_csv(filepath, header=7)
df.columns = ['ch', 'n']

ch = df['ch']  # bin
n = df['n'] / duration - bg_n  # counts/s
dn = np.sqrt(n)

plt.plot(get_energy(ch), n, 'k')
# plt.plot(ch, n + dn, '--k', alpha=0.5)
plt.xlim(0, 500)
plt.xlabel('Energy [keV]')
plt.ylabel('Counts/s')
plt.yscale('log')
plt.title('Air (Short sample)')

# Look for Pb peak at 351.9 keV
lower_idx = 325
upper_idx = 375

y = n[lower_idx:upper_idx]
amax = np.amax(y)
index = y.argmax() + lower_idx

plt.axvspan(ch[lower_idx], ch[upper_idx], color=c[0], alpha=0.3)
# plt.axhline(amax, color=c[0])
# plt.axvline(index, color=c[0])

bq_meas1 = amax
dbq_meas1 = np.sqrt(amax)

print(f"Peak position: {index} keV")
print(f"Counts/s: {bq_meas1:.2f} +/- {dbq_meas1:.2f}")
```

## Spectrum 2: long sampling time

```python
filepath = 'data/air_long.csv'
duration = 1800  # s

df = pd.read_csv(filepath, header=7)
df.columns = ['ch', 'n']

ch = df['ch']  # bin
n = df['n'] / duration - bg_n  # counts/s
dn = np.sqrt(n)

plt.plot(get_energy(ch), n, 'k')
# plt.plot(ch, n + dn, '--k', alpha=0.5)
plt.xlim(0, 900)
plt.xlabel('Energy [keV]')
plt.ylabel('Counts/s')
plt.yscale('log')
plt.title('Air (Long sample)')
```
